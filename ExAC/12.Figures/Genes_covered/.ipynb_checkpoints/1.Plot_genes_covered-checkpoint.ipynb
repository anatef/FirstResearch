```r
library(ggplot2)
library(scales)
library(ggpubr)

#Number of genes total from Ensembl
#Number of genes covered is calculated in: 5.domains_stats/domains_statistics.ipynb
#Number of solved structures was calculated in PDB on Dec 19, 2016.

genes_num <- c(20805, 19823, 13254, 5566)
types <- c("All human\ngenes", "contain\ndomains", "frequent (>10)\ndomains", "solved\nstructure")
percentile <- c("100%", "95.3%", "63.7%", "26.7%")
per_vjust <- c(14.5,14,9,4.25)
fill_colors <- c("#252525", "#525252", "#969696", "#bdbdbd")
                 
genes_covered <- data.frame(genes_num, types, percentile, per_vjust, fill_colors, stringsAsFactors=TRUE)


final_fig <- ggplot(genes_covered, aes(x=types, y=genes_num, fill=types)) +
  geom_bar(stat='identity', position = 'dodge') +
  geom_text(aes(label=comma(genes_num)), position=position_dodge(width=0.9), vjust=-0.25, size=5) +
  geom_text(aes(label=percentile), position=position_dodge(width=0.9), vjust=per_vjust, color="white", size=7) +
  theme(axis.text=element_text(size=4)) +
  scale_x_discrete(limits=types) +
  scale_y_continuous(labels = comma) +
  scale_fill_manual(values=fill_colors) +
  guides(fill=FALSE) +
  xlab("") +
  ylab("Number of genes covered") +
  theme_bw() +
  theme(
    axis.text.x=element_text(size=14),
    axis.text.y=element_text(size=14),
    axis.title.y=element_text(size=18)
  )

options(repr.plot.width=8, repr.plot.height=10)
svg("plots/genes_covered.svg", width=8, height=10)
pdf("plots/genes_covered.pdf", width=8, height=10)
final_fig
dev.off()
final_fig
```
