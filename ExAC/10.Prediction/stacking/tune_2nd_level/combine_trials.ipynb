### Combine files across trials for every ligand+classifier+fold combination

```python
for ligand in ligands:
    
    for classifier in classifiers:
        for i in range(folds_num):
        #for i in range(1,2):
            fold = i+1
            
            print "fold="+str(fold)
            fold_df = pd.DataFrame()
            for trial in range(no_trials):
                try:
                    trial_filename = input_path+ligand+"_"+classifier+"_fold"+str(fold)+"_trial"+str(trial)+"_"+str(folds_num)+"w_hyperparameters.csv"
                    trial_hyperparameters_df = pd.read_csv(trial_filename, sep='\t', index_col=0)
                    fold_df = fold_df.append(trial_hyperparameters_df)
                except:
                    continue
            
            fold_global_auprc = fold_df.sort_values(by = "mean_AUPRC", ascending = False)
            fold_global_auprc.to_csv(output_path+ligand+"_"+classifier+"_fold"+str(fold)+"_mean_AUPRC_hyperparams.csv", sep='\t')
            
            #fold_domain_auprc_ratio = fold_df.sort_values(by = "mean_dom_AUC", ascending = False)
            #fold_domain_auprc_ratio.to_csv(output_path+ligand+"_"+classifier+"_fold"+str(fold)+"_mean_dom_AUC_hyperparams.csv", sep='\t')
            
            fold_test_auprc = fold_df.sort_values(by = "test_AUPRC", ascending = False)
            fold_test_auprc.to_csv(output_path+ligand+"_"+classifier+"_fold"+str(fold)+"_test_AUPRC_hyperparams.csv", sep='\t')
            
            #fold_test_dom_auprc_ratio = fold_df.sort_values(by = "test_dom_AUC", ascending = False)
            #fold_test_dom_auprc_ratio.to_csv(output_path+ligand+"_"+classifier+"_fold"+str(fold)+"_test_dom_AUC_hyperparams.csv", sep='\t')
header = fold_global_auprc.columns.tolist()
```
