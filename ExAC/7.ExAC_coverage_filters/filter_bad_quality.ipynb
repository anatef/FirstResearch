```python
import pandas as pd
import numpy as np
import pickle
import datetime
from collections import defaultdict
from IPython.core.display import HTML
HTML("<style>.container { width:100% !important; }</style>")
```

```python
curr_dir = !pwd
domains_th = "10"
input_path = curr_dir[0]+"/../6.Ext_features_to_dicts/ext_features_dicts/pfam-v30/"

#Read the list of domains
with open(curr_dir[0]+"/../5.domains_stats/filtered"+domains_th+"_list.pik", 'rb') as handle:
    filtered_domains_list = pickle.load(handle)
filtered_domains_list.sort()
```

### Filter by coverage

```python
%%time
coverage_filtered_dict = []
for domain_name in filtered_domains_list:
    
    dirfiles = !ls -t $input_path$domain_name
    filename = dirfiles[0]
    with open(input_path+domain_name+"/"+filename, 'rb') as handle:
        states_dict = pickle.load(handle)
    
    cov_states_dict = defaultdict(list)
    coverage_filtered_dict = defaultdict(list)
    for state in states_dict.keys():
        
        for d in states_dict[state]:
            mean_cov = d["coverage_mean"]
            
            if (mean_cov < 50):
                coverage_filtered_dict[state].append(d)
            
            cov_states_dict[state].append(d)
    
    #Saving the updated dictionary
    !mkdir -p filtered_dicts/pfam-v30/$domain_name
    
    with open(curr_dir[0]+"/filtered_dicts/pfam-v30/"+domain_name+"/"+domain_name+"_filtered_states_dict_"+datetime.date.today().strftime("%m.%d.%y")+".pik", 'wb') as handle:
        pickle.dump(cov_states_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
    with open(curr_dir[0]+"/filtered_dicts/pfam-v30/"+domain_name+"/"+domain_name+"_filtered_out_dict_"+datetime.date.today().strftime("%m.%d.%y")+".pik", 'wb') as handle:
        pickle.dump(coverage_filtered_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
    print "Finished "+domain_name      
```
