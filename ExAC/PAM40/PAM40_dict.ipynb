```python
import pandas as pd
from collections import defaultdict
import fileinput
import pickle
curr_dir = !pwd

pam40_dict = defaultdict(dict)

for line in fileinput.input(curr_dir[0]+"/PAM40.txt"):  
    #Reading aa order
    if (line[0] == " "):
        aa_order = line.split()
        continue
    
    #Reading values to dict
    vals = line.split()
    for i in range(1,len(vals)):
        pam40_dict[vals[0]][aa_order[i-1]] = int(vals[i])

fileinput.close()

#Saving to file
with open(curr_dir[0]+'/PAM40_dict.pik', 'wb') as handle:
    pickle.dump(pam40_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
