```python
#Reading the dictionary data file
curr_dir = !pwd
domain_name = "zf-C2H2"
input_path = curr_dir[0]+"/../5.HMM_alter_align/domains_states_dicts/pfam-v30/"+domain_name+"/"
dirfiles = !ls -t $input_path
filename = dirfiles[0]

with open(input_path+filename, 'rb') as handle:
    states_dict = pickle.load(handle)

#Getting the number of states for this domain
states_num = len(states_dict.keys())
print states_num
```

```python
with open(curr_dir[0]+"/../2.parse_Pfam/v30/domains_hmm_prob_dict.pik", 'rb') as handle:
    hmm_prob_dict = pickle.load(handle)
```

```python
with open(curr_dir[0]+"/../binding_score/domains_all_binding_dict.pik", 'rb') as handle:
    binding_scores_dict = pickle.load(handle)
```

```python
"Ank_3" in binding_scores_dict.keys()
```

```python
#Thresholds for conserved states:
#con_thresholds = {"ig": 0.5, "zf-C2H2": 0.5, "EGF": 0.5, "Ank": 0.5, "KRAB": 0.5, "Sushi": 0.5}


domain_prob = hmm_prob_dict[domain_name]
con_list = []
for state in domain_prob.keys():
    for x in domain_prob[state]:
        if (x > 0.5):
            con_list.append(state)
len(con_list)
```

```python
con_list
```

```python
#Add JSD to dict
for state in states_dict.keys():
    for d in states_dict[state]:
        if (len(d["alterations_af_adj_dict"].keys()) == 0):
            d["JSD_adj"] = 0
        else:
            if (d["aa_ref"] == "X"):
                d["aa_ref"] = "*"
            d["JSD_adj"] = JSD(d["alterations_af_adj_dict"], d["aa_ref"], d["af_adj"], state, JSD_background.MAJOR_ALLELE)
```

```python
#Create a dict of average JSD and binding scores
avg_jsd_binding_dict = {}
for state in states_dict.keys():
    jsd_sum = 0
    jsd_num = len(states_dict[state])
    for d in states_dict[state]:
        jsd_sum += d["JSD_adj"]
    avg = -(jsd_sum/float(jsd_num)) #Converting JSD to negative
    avg = (1+avg)
    avg_jsd_binding_dict[state] = [avg]

for state in range(1,states_num+1):
    if (state in binding_scores_dict[domain_name]["states"]):
        score_idx = np.where(np.array(binding_scores_dict[domain_name]["states"]) == state)[0]
        max_bind_score = 0
        for i in score_idx:
            bind_score = binding_scores_dict[domain_name]["scores"][i]
            if (bind_score > max_bind_score):
                max_bind_score = bind_score
    else:
        max_bind_score = 0.0
    avg_jsd_binding_dict[state].append(max_bind_score)
```

```python
#Create a dict of non-conserved states avg JSD
non_con_avg_jsd_dict = {}
for state in avg_jsd_binding_dict.keys():
    if (state in con_list):
        continue
    non_con_avg_jsd_dict[state] = avg_jsd_binding_dict[state]
```

```python
with open(curr_dir[0]+"/data-frames/avg_JSD/"+domain_name+"_avg_JSD_neg_df.csv", 'w') as f:
    f.write("state,avg_JSD,bind_score\n")
    [f.write('{0},{1},{2},{3}\n'.format(key, key, value[0], value[1])) for key, value in avg_jsd_binding_dict.items()]
```

```python
with open(curr_dir[0]+"/data-frames/avg_JSD/"+domain_name+"_noncon_avg_JSD_neg_df.csv", 'w') as f:
    f.write("state,avg_JSD,bind_score\n")
    [f.write('{0},{1},{2},{3}\n'.format(key, key, value[0], value[1])) for key, value in non_con_avg_jsd_dict.items()]
```

```python
### Creating flat dictionaries
#Create a dictionary for AF values only
states_af_dict = defaultdict(list)
for state in states_dict.keys():
    for d in states_dict[state]:
        states_af_dict[state].append(d["af"])

#Create a dictionary for AF adjusted values only
states_af_adj_dict = defaultdict(list)
for state in states_dict.keys():
    for d in states_dict[state]:
        states_af_adj_dict[state].append(d["af_adj"])
```

```python
#Create a dictionary for Coverage values only
states_mean_cov_dict = defaultdict(list)
for state in states_dict.keys():
    for d in states_dict[state]:
        states_mean_cov_dict[state].append(d["coverage_mean"])
```

```python
#Scaling the MAF
states_MAF_adj_dict_scaled = defaultdict(list)
for state in states_dict.keys():
    state_len = len(states_dict[state])
    for d in states_dict[state]:
        states_MAF_adj_dict_scaled[state].append(float(d["af_adj"]/state_len))
```

```python
#Create a dictionary for af adj, chrom, chrom_pos
states_MAF_adj_chrom_pos_dict = defaultdict(list)
for state in states_dict.keys():
    state_len = len(states_dict[state])
    for d in states_dict[state]:
        states_MAF_adj_chrom_pos_dict[state].append((d["af_adj"], d["JSD_adj"], d["chrom"], d["chrom_pos"][0]))
```

### Trim 0s before calculating EMD

Find the lowest number of 0s in a distribution. Remove this number of 0s from all distriubtions. Then, calculate EMD and try to see boxplots and medians.
This isn't good for cases usch as state #12: high number of non-zero low values.

```python
#Count the number of zero Adj MAF
zero_adj_af = [0] * states_num
for state in states_dict.keys():
    for d in states_dict[state]:
        if (d["af_adj"] == 0):
            zero_adj_af[state-1] += 1
```

```python
min_0s = min(zero_adj_af)
```

```python
#Trim each state by the same number of data points
states_dict_0s = states_dict.copy()
for state in states_dict_0s.keys():
    #Sort the state list
    state_sorted = sorted(states_dict_0s[state], key=lambda k: k['af_adj'])
    states_dict_0s[state] = state_sorted[min_0s:]
```

```python
maf_med = []
for state in states_dict_0s.keys():
    mafs = []
    for d in states_dict_0s[state]:
        mafs.append(d["af_adj"])
    maf_med.append(np.median(mafs))
```

```python
#Create a dictionary for MAF adj values only
states_MAF_adj_dict_0s = defaultdict(list)
for state in states_dict_0s.keys():
    for d in states_dict_0s[state]:
        states_MAF_adj_dict_0s[state].append(d["af_adj"])
```

```python
trim0s_maf_emd_dict = compute_pymed_emd(states_MAF_adj_dict_0s, 1000)
plot_emd(trim0s_maf_emd_dict, "Adj MAF - trim 0s", curr_dir, "SNP_Zinc_filter_JSD_Major_allele_EMD", True, False)
```

### Trim 90% of 0s before EMD

```python
#Count the number of zero Adj MAF
zero_adj_af = [0] * states_num
for state in states_dict.keys():
    for d in states_dict[state]:
        if (d["af_adj"] == 0):
            zero_adj_af[state-1] += 1
```

```python
#Trim each state by the same % of 0 data points
trim_per = 0.9

states_dict_0s_per = states_dict.copy()
for state in states_dict_0s.keys():
    #Sort the state list
    state_sorted = sorted(states_dict_0s_per[state], key=lambda k: k['af_adj'])
    trim_num = int(np.round(zero_adj_af[state-1] * trim_per))
    states_dict_0s_per[state] = state_sorted[trim_num:]
```
