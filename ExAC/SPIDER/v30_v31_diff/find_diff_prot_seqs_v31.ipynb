### What about domains with genes and a different prot id? - correction for previous code!!
1. Check for domains that changed prot id for a particular gene

2. A loop that goes over all the domains canonic prot and checks what are the differences between the versions.

```python
new_prot_dict = defaultdict(dict)
new_genes_dict = defaultdict(dict)
updated_domain_gene_sequence = defaultdict(dict)
for domain_name in v31_domains:
    if (domain_name in v31_new_domains):
        continue
    #print "new_domain = "+domain_name
    filename = domain_name+".csv"
    in_path31 = curr_dir[0]+"/../../3.parse_HMMER/hmm_domains/pfam-v31/"
    
    domain_data31 = pd.read_csv(in_path31+filename, sep='\t', index_col=0, dtype={"chrom_num": str})
    
    #Sort the domain data
    sorted_domain_data31 = domain_data31.sort_values(by=["chrom_num", "gene", "TargetStart"])
    sorted_domain_data31 = sorted_domain_data31.reset_index(drop=True)
    
    in_path30 = curr_dir[0]+"/../../3.parse_HMMER/hmm_domains/pfam-v30/"
    domain_data30 = pd.read_csv(in_path30+filename, sep='\t', index_col=0, dtype={"chrom_num": str})

    #Sort the domain data
    sorted_domain_data30 = domain_data30.sort_values(by=["chrom_num", "gene", "TargetStart"])
    sorted_domain_data30 = sorted_domain_data30.reset_index(drop=True)
    
    #Get the canonic protein ids file for the domain
    with open(curr_dir[0]+"/../../4.parse_Uniprot/domains_canonic_prot/pfam-v31/"+domain_name+"_canonic_prot.pik", 'rb') as handle:
        canonic_protein31 = pickle.load(handle)
    with open(curr_dir[0]+"/../../4.parse_Uniprot/domains_canonic_prot/pfam-v30/"+domain_name+"_canonic_prot.pik", 'rb') as handle:
        canonic_protein30 = pickle.load(handle)
    
    genes30 = sorted_domain_data30["gene"].unique()
    genes31 = sorted_domain_data31["gene"].unique()
    
    for gene in genes31:
        
        chrom = sorted_domain_data31[sorted_domain_data31["gene"] == gene]["chrom_num"].unique()[0]
        if (chrom not in chromosome_names):
            continue
        
        #Check if the gene is new for the domain
        prot_id31 = canonic_protein31[gene]
        if (gene not in genes30):
            new_genes_dict[domain_name][gene] = prot_id31
            updated_domain_gene_sequence[domain_name][gene] = prot_dict_v31[gene][prot_id31]
            continue
            
       #Check if the gene is known but the prot id is new for the gene
        prot_id30 = canonic_protein30[gene]
        if (prot_id31 != prot_id30):
            new_prot_dict[domain_name][gene] = prot_id31
            updated_domain_gene_sequence[domain_name][gene] = prot_dict_v31[gene][prot_id31]
    print "finished domain "+domain_name
        
```

```python
with open(curr_dir[0]+"/v30-v31_diff_domains_genes_prot_seq.pik", 'wb') as handle:
    pickle.dump(updated_domain_gene_sequence, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

### Write SPIDER results to dict and add it to the existing domain dict

```python
def read_file(filepath, output_dict, gene_id):
    # Differentiate between files with same feature names
    tokens = filepath.split('.')
    extension = tokens[len(tokens)-1]

    # Read files and save to dict
    with open(filepath,'r') as f:
        for line in f.readlines():
            # Parse header
            if line[0] == '#':
                tokens = line.split()
                features = tokens[1:len(tokens)]
                continue

            # Parse lines
            tokens = line.split()
            if not int(tokens[0]) in output_dict[gene_id].keys():
                output_dict[gene_id][int(tokens[0])] = {}
            for i in range(0,len(features)):
                output_dict[gene_id][int(tokens[0])][extension+"_"+features[i]] = tokens[i+1]
```
